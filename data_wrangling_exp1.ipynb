```python
import pandas as pd
import numpy as np
```

```python
# Read the employees.csv file into a DataFrame

df = pd.read_csv("./data/employees.csv")
df.head()
```

```python
# Inspect column dtypes and non-null counts

df.info()
```

```python
# Number of null values in each column

df.isnull().sum()
```

```python
# Identify which columns are numeric vs non-numeric

numeric_cols = df.select_dtypes(include='number').columns
non_numeric_cols = df.select_dtypes(exclude='number').columns

print("Numeric columns:", list(numeric_cols))
print("Non-numeric columns:", list(non_numeric_cols))
```

```python
# Drop rows that have a null value in any non-numeric column

df2 = df.dropna(subset=non_numeric_cols)
df2.isnull().sum()
```

```python
# Summary stats (count, mean, std, min, max, etc.) for the numeric columns

summary = df2[numeric_cols].describe()
summary
```
